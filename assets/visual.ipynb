```python
import json
from google.colab import files

# Subir el archivo GeoJSON
uploaded = files.upload()

# Cargar el archivo GeoJSON
with open('Spain.json', 'r') as f:
    geojson_data = json.load(f)

# Verificar que el archivo se cargó correctamente
print(geojson_data['features'][1]['properties'])

```

```python
community_name_mapping = {
    'Andalucía': 'Andalucía',
    'Aragón': 'Aragón',
    'Asturias, Principado de': 'Asturias',
    'Balears, Illes': 'Islas Baleares',
    'Canarias': 'Canary Is.',
    'Cantabria': 'Cantabria',
    'Castilla y León': 'Castilla y León',
    'Castilla - La Mancha': 'Castilla-La Mancha',
    'Cataluña': 'Cataluña',
    'Comunitat Valenciana': 'Valenciana',
    'Extremadura': 'Extremadura',
    'Galicia': 'Galicia',
    'Madrid, Comunidad de': 'Madrid',
    'Murcia, Región de': 'Murcia',
    'Navarra, Comunidad Foral de': 'Foral de Navarra',
    'País Vasco': 'País Vasco',
    'Rioja, La': 'La Rioja',
    'Ceuta': 'Ceuta',
    'Melilla': 'Melilla',
    'Total Nacional': None
}

df['Comunidad'] = df['Comunidades y Ciudades Autónomas'].map(community_name_mapping)

# Filtra valores no mapeados o nulos
df = df[df['Comunidad'].notnull()]

# Filtra los datos para el año 2023 y sexo "Ambos sexos"
df['Periodo'] = df['Periodo'].astype(str).str.strip()
df_2023 = df[(df['Periodo'] == '2023') & (df['Sexo'] == 'Ambos sexos')]

# Agrupa los datos por comunidad autónoma
df_grouped = df_2023.groupby('Comunidad').agg({'Total': 'mean'}).reset_index()

# Crear el mapa coroplético
map_chart = alt.Chart(
    alt.Data(values=geojson_data, format={'property': 'features', 'type': 'json'})
).mark_geoshape().encode(
    color=alt.Color(
        'Total:Q',
        scale=alt.Scale(
            scheme='viridis',  # Mantiene la paleta original
            reverse=True       # Invierte el mapeo de los valores
        ),
        title='Tasa de Empleo (%)'
    ),
    tooltip=['properties.name:N', 'Total:Q']
).transform_lookup(
    lookup='properties.region',  # Campo en el GeoJSON que corresponde a la región
    from_=alt.LookupData(df_grouped, 'Comunidad', ['Total'])  # Campo en el DataFrame
).properties(
    title='Tasa de empleo por comunidad autónoma en 2023 (Ambos sexos)',
    width=800,
    height=600
)

map_chart.display()
```

```python
# Filtra los datos para el año 2023 y para ambos sexos
df['Periodo'] = df['Periodo'].astype(str).str.strip()
df_2023 = df[(df['Periodo'] == '2023') & (df['Sexo'].isin(['Hombres', 'Mujeres']))]  # Filtra por ambos sexos

# Agrupa los datos por comunidad y sexo (hombres y mujeres)
df_grouped = df_2023.groupby(['Comunidad', 'Sexo']).agg({'Total': 'mean'}).reset_index()

# Pivotamos el DataFrame para tener las tasas de empleo por comunidad para hombres y mujeres
df_pivoted = df_grouped.pivot(index='Comunidad', columns='Sexo', values='Total').reset_index()

# Asignar un color según cuál sexo tiene una tasa de empleo mayor
df_pivoted['Predomina'] = df_pivoted.apply(
    lambda row: 'Hombres' if row['Hombres'] > row['Mujeres'] else 'Mujeres', axis=1
)

# Asignar colores: azul para hombres, rosa para mujeres
df_pivoted['Color'] = df_pivoted['Predomina'].apply(lambda x: 'blue' if x == 'Hombres' else 'pink')

# Crear el mapa coroplético con color basado en si predominan los hombres o las mujeres
map_chart = alt.Chart(
    alt.Data(values=geojson_data, format={'property': 'features', 'type': 'json'})
).mark_geoshape().encode(
    color=alt.Color('Predomina:N', scale=alt.Scale(domain=['Hombres', 'Mujeres'], range=['blue', 'pink']), title='Predomina Sexo'),
    tooltip=['properties.name:N', 'Predomina:N', 'Hombres:Q', 'Mujeres:Q']
).transform_lookup(
    lookup='properties.region',  # Campo en el GeoJSON que corresponde a la región
    from_=alt.LookupData(df_pivoted, 'Comunidad', ['Predomina', 'Hombres', 'Mujeres'])  # Campos en el DataFrame
).properties(
    title='Predominio de la tasa de empleo entre hombres y mujeres por comunidad autónoma en 2023',
    width=800,
    height=600
)

map_chart.display()
```

```python


# Asegurarse de que 'Periodo' sea texto y limpio
df['Periodo'] = df['Periodo'].astype(str).str.strip()

# Filtrar datos para el año 2023 y eliminar "Total Nacional"
df_2023 = df[(df['Periodo'] == '2023') &
             (df['Comunidades y Ciudades Autónomas'] != 'Total Nacional') &
             (df['Sexo'] != 'Ambos sexos')]

# Separar los datos para mujeres y hombres
df_mujeres = df_2023[df_2023['Sexo'] == 'Mujeres']
df_hombres = df_2023[df_2023['Sexo'] == 'Hombres']

# Hacer merge de los dos DataFrames para tener las tasas de empleo de mujeres y hombres juntos
df_merged = pd.merge(df_mujeres[['Comunidades y Ciudades Autónomas', 'Total']],
                     df_hombres[['Comunidades y Ciudades Autónomas', 'Total']],
                     on='Comunidades y Ciudades Autónomas',
                     suffixes=('_Mujeres', '_Hombres'))

# Calcular la diferencia de la tasa de empleo entre hombres y mujeres
df_merged['Wiferencia'] = df_merged['Total_Hombres'] - df_merged['Total_Mujeres']

# Reformatear el DataFrame para Altair
df_merged_melted = df_merged.melt(id_vars=['Comunidades y Ciudades Autónomas'],
                                  value_vars=['Total_Mujeres', 'Wiferencia'],
                                  var_name='Sexo', value_name='Tasa de Empleo')

# Ajustar los valores de 'Sexo' para el orden correcto
df_merged_melted['Sexo'] = df_merged_melted['Sexo'].map({
    'Total_Mujeres': 'Mujeres',
    'Wiferencia': 'Mujeres y Hombres'
})

# Ajustar el orden de las barras para que las mujeres se muestren primero
df_merged_melted['Sexo'] = pd.Categorical(df_merged_melted['Sexo'], categories=['Mujeres', 'Mujeres y Hombres'], ordered=True)

# Crear el gráfico de barras apiladas, donde las mujeres están al frente
bar_chart = alt.Chart(df_merged_melted).mark_bar().encode(
    x=alt.X('Tasa de Empleo:Q',
            scale=alt.Scale(domain=[0, 100]),  # Escala desde 0 hasta 100
            title='Tasa de Empleo (%)'),
    y=alt.Y('Comunidades y Ciudades Autónomas:N',
            sort='-x',  # Ordenar por la tasa de empleo total
            title='Comunidades Autónomas'),
    color=alt.Color('Sexo:N',
                    scale=alt.Scale(domain=['Mujeres', 'Mujeres y Hombres'],
                                    range=['pink', 'blue']),  # Colores personalizados
                    title='Sexo'),
    tooltip=['Comunidades y Ciudades Autónomas', 'Sexo', 'Tasa de Empleo'],  # Información del tooltip
).properties(
    title='Tasa de empleo por comunidad autónoma en 2023 (Mujeres y diferencia con Hombres)',
    width=600,
    height=600
)

bar_chart.display()
```
